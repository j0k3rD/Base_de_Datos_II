## Ejercicio 1

- Descomprimir el archivo clima.tgz de la carpeta data
- Convertir cada archivo a Dataframe y mostrarlo en pantalla

```python
# Importar archivo tar
import tarfile

# Abrir archivo tar.gz tgz
objetoTar = tarfile.open("../data/clima.tgz")
# Extraer todo el contenido en la carpeta
objetoTar.extractall("../data/clima-tp1ejer1")
#Obtengo los nombres de los archivos
file_names = objetoTar.getnames()
# Cerrar lectura
objetoTar.close()
tables = []

#Pasar a df todos los archivos .txt
import pandas as pd

# Transformar archivos de .txt a Dataframe y los mostramos
for objeto in file_names:
    archivo = "../data/clima-tp1ejer1/" + objeto
    df = pd.read_csv(archivo, delimiter = "\t", encoding='latin-1', header=[0,1])
    print(df)
```

## Ejercicio 2

- Obtener de la API de <a href="https://wikimedia.org/api/rest_v1/#/Pageviews%20data">Wikimedia </a> la cantidad de visitas en este año a la página Albert Einstein por día.
- Mostrar el Dataframe obtenido por pantalla

```python
import requests
import pandas as pd

#Obtenemos URL con la api rest de Wiki
url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/es.wikipedia.org/all-access/all-agents/Albert_Einstein/daily/20210101/20221230"

#Obtenemos el modelo
r = requests.get(url, headers={'user-agent':'Mozilla/5.0 (X11; Ubuntu; Linux i686; rv:91.0) Gecko/20100101 Firefox/91.0'})
#Transformamos a json
data = r.json()

df = pd.DataFrame(data["items"])
df
```

## Ejercicio 4

- Obtener de la API <a href="https://date.nager.at/swagger/index.html">Nager.Date</a> las fiestas nacionales argentinas para el 2021
- Mostrar el dataframe resultante por pantalla

```python
# Importar request y pandas
import requests
import pandas as pd

# URL de la consulta.
url = "https://date.nager.at/api/v3/PublicHolidays/2021/AR"

# Realizamos la request y transformamos a json.
req = requests.get(url, headers={'user-agent':'Mozilla/5.0 (X11; Ubuntu; Linux i686; rv:91.0) Gecko/20100101 Firefox/91.0'}) 
data = req.json()

# Transformamos a tabla y mostramos.
df = pd.DataFrame(data)
df
```
