## Ejemplo 1

<img src=https://www.consultantsmind.com/wp-content/uploads/2017/04/Consultantsmind-United-Airlines-departures.png>

Respuesta: Es una mala representación, porque al FY16 lo muestra un "x3" mas gránde que los dos valores de la izquierda, cuando de un 56% a un 63% hay un diferencia muy pequeña. Esto lo realizan para dar una imagen erronea de la diferencia de ambos, haciendo creer que la diferencia es muy grande cuando en realidad no lo es.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [u'FY14',u'FY15',u'FY16']
y = [55.2, 56.0, 63.6]
# Cambiar tamaño y forma de barras
fig, ax = plt.subplots()    
width = 0.75 # cambiar ancho de barras 
ind = np.arange(len(y))
ax.barh(ind, y, width, color="blue")
ax.set_yticks(ind+width/30)
ax.set_yticklabels(x, minor=False)

# Cambiar tamaño de visualización
plt.xlim([30, 70])

# Mostrar valores al final de la barra.
for i, v in enumerate(y):
    ax.text(v + 1, i - .03, str(str(v) + "%"), color='blue', fontweight='bold')
plt.title("Departamento de United Airlines")
plt.xlabel('%')
plt.ylabel('FY')    
plt.show()
```

Respuesta: Es una mala representación, porque no se entiende cual es el que tiene mayor valor de manera visual, ademas que la suma de todos no da un 100%, siendo poco intuitivo el ver las diferencias. Lo hago mediatne gráfico de barra, para notar mejor las diferencias de ambos.

```python
import matplotlib.pyplot as plt

# Datos
x = ['Rahmat Shah', 'Kusal Perera', 'Root', 'Pooran', 'Du Plessis', 'Babar', 'Warner', 'Sahkib', 'Rohit', 'Williamson']
y = [14.8, 18.16, 19.07, 20.01, 21.06, 24.51, 25.02, 28.25, 29.05, 30.23] 

# Creación del gráfico
plt.figure(figsize=(13, 5)) 
plt.bar(x,y)
plt.xlabel("Jugadores")
plt.ylabel("% de corridas anotadas por el máximo goleador")
plt.title("Grandes anotadas de la copa del mundo")
plt.show();
```

## Ejemplo 3

<img src=https://www.consultantsmind.com/wp-content/uploads/2017/04/Consultantsmind-Facebook-Revenue-per-User.png>

Respueta: La comparación entre ambos gráficos, por el tamaño hace parecer que son parecidos, pero por el valor son muy diferentes ambos. Ademas de que el gráfico de barras superpuestos no es muy entendible, por lo que lo separo en diferentes barras para que quede mas claro.

```python
# Datos Worldwide
x = ["Q3'15", "Q4'15", "Q1'16", "Q2'16", "Q3'16"]
y = [2.83, 3.60, 3.21, 3.71, 3.89]
x2 = [0.5,1.5,2.5,3.5,4.5]
y2 = [2.97, 3.73, 3.32, 3.82, 4.01] 

# Creación del gráfico Worldwide
plt.bar(x, y ,label="A",color="c",width=.5)
plt.bar(x2, y2, label="B", color="r",width=.5)
plt.legend()
plt.ylabel("$")
plt.ylim(0, 16)
plt.title("Worldwide")
plt.show()

# Datos Us & Canada
x = ["Q3'15", "Q4'15", "Q1'16", "Q2'16", "Q3'16"]
y = [9.86, 13.06, 11.86, 13.74, 15.08]
x2 = [0.5,1.5,2.5,3.5,4.5]
y2 = [10.49, 13.70, 12.43, 14.34, 15.65] 

# Creación del gráfico Us & Canada
plt.bar(x, y ,label="A",color="c",width=.5)
plt.bar(x2, y2, label="B", color="r",width=.5)
plt.legend()
plt.ylabel("$")
plt.title("Us & Canada")
plt.show()
```
