Respuesta: Es una mala representación, porque al FY16 lo muestra un "x3" mas gránde que los dos valores de la izquierda, cuando de un 56% a un 63% hay un diferencia muy pequeña. Esto lo realizan para dar una imagen erronea de la diferencia de ambos, haciendo creer que la diferencia es muy grande cuando en realidad no lo es.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [u'FY14',u'FY15',u'FY16']
y = [55.2, 56.0, 63.6]
# Cambiar tamaño y forma de barras
fig, ax = plt.subplots()    
width = 0.75 # cambiar ancho de barras 
ind = np.arange(len(y))
ax.barh(ind, y, width, color="blue")
ax.set_yticks(ind+width/30)
ax.set_yticklabels(x, minor=False)

# Cambiar tamaño de visualización
plt.xlim([30, 70])

# Mostrar valores al final de la barra.
for i, v in enumerate(y):
    ax.text(v + 1, i - .03, str(str(v) + "%"), color='blue', fontweight='bold')
plt.title("Departamento de United Airlines")
plt.xlabel('%')
plt.ylabel('FY')    
plt.show()
```
