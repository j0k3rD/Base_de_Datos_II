```python
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plp
%matplotlib inline
import seaborn as sns

df = pd.read_csv(r'data/sponge.data', delimiter = ",")
df
```

```python
#verificamos el tipo de elemento
df.dtypes
```

```python
#seleccionamos los faltantes, reemplazando los "?", utilizamos este metodo al encontrar missing values en Python.
faltantes =[]
#buscamos por columna
for column in df.columns:
    item = {}
    #igualamos las columnas
    item["column"] = column
    #pasamos a df el tipo como dato
    item["type"] = df[column].dtypes
    #sacamos el porcentaje de la columna faltante x100
    item["percentage"] = np.mean(df[column] == "?")*100
    #sumamos la columna cantidad faltante
    item["quantity"] = np.sum(df[column] == "?")
    #agregamos los cambios
    faltantes.append(item)

#creamos el dataframe nuevo con pandas
faltantes_df = pd.DataFrame(faltantes)
#llamamos
faltantes_df
```

```python
#agrupamos los datos faltantes que tengas porcentaje mayos a 0
missing_df = faltantes_df.loc[faltantes_df["percentage"] > 0]
missing_df
```

```python
import pandas as pd
import numpy as np
import requests
import io

with requests.get(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/echocardiogram/echocardiogram.data"
) as url:
    df = pd.read_csv(io.StringIO(url.content.decode("utf-8")),
                     sep="/n", header=None)
    df = df[0].str.split(",", expand=True)
df
```

```python
#imputamos los valores perdidos, utilizamos este metodo por la gran cantidad de datos.
faltantes = []

for column in df.columns:
    item = {}
    item["column"] = column
    #transformamos la columna a dtype
    item["type"] = df[column].dtypes
    #hacemos el porcentaje de la columna nula
    item["percentage"] = np.mean(df[column].isnull())*100
    #agrupamos
    item["cuantity"] = np.sum(df[column].isnull())
    #agregamos los cambios
    faltantes.append(item)

#creamos de df y llamamos
faltantes_df = pd.DataFrame(faltantes)
faltantes_df
```

```python
#copiamos df sin reemplazar los objetos (deep)
df_to_clear = df.copy(deep = False)
#droopeamos las filas y columnas
clear_df = df_to_clear.drop(13, axis = 1)
clear_df
```

```python
faltantes = []

for column in clear_df.columns:
    item = {}
    item["column"] = column
    item["type"] = clear_df[column].dtypes
    item["percentage"] = np.mean(clear_df[column] == '?')*100
    item["quantity"] = np.sum(clear_df[column] == '?')
    faltantes.append(item)

faltantes_df = pd.DataFrame(faltantes)
faltantes_df
```

```python
import matplotlib.pyplot as plt

#creamos el grafico de barras
plp.bar(faltantes_df["column"],faltantes_df["quantity"])
#eje y
plp.ylabel('Quantity')
#eje x
plp.xlabel('Columns')
plp.show()
```
